```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/datasets/presidents.csv')
df.head()
```

```python
df["First"] = df['President']
df["First"] = df["First"].replace("[ ].*", "", regex=True)
df.head()
```

```python
del (df['First'])
```

```python
def splitname(row):
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    return row

df = df.apply(splitname, axis='columns')
df.head()
```

```python
del(df['First'])
del(df['Last'])
```

```python
pattern="(^[\w]*)(?:.* )([\w]*$)"

df['President'].str.extract(pattern)
```

```python
pattern == "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names = df['President'].str.extract(pattern)
names
```

```python
df["First"] = names["First"]
df["Last"] = names["Last"]
df.head()
```

```python
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

# Extract names using the pattern
names = df['President'].str.extract(pattern)

# Reset the index of names to match df's index
names = names.set_index(df.index)

# Assign the extracted columns to df
df["First"] = names["First"]
df["Last"] = names["Last"]
df.head()
```

```python
df["Born"] = df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

```python
df["Born"] = pd.to_datetime(df["Born"])
df["Born"].head()
```
